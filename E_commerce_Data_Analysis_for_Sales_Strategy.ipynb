# **1. Sales_data klasöründe her aya ait satış verileri bulunmaktadır. Bütün aylara ait satış verilerini bilgileri kaydetmeden bir dataframe de toplayacağız:** 

```python
import numpy as np 
import pandas as pd
import os
import matplotlib as plt
```

```python
files = [file for file in os.listdir('/content/all_data')]

all_months_data = pd.DataFrame()

for file in files:
    df = pd.read_csv('/content/all_data/'+ file)
    all_months_data = pd.concat([all_months_data, df])

all_months_data.head()

```

```python
all_months_data.to_csv("all_data.csv", index=False) # Tek dosyada topladık ve kaydettik
```

```python
all_data = pd.read_csv("all_data.csv")
all_data.head()
```

# **2. Boş değer (null) içeren satırları veri setinden çıkarılması:**

```python
nan_df = all_data[all_data.isna().any(axis=1)]
nan_df.head()

all_data = all_data.dropna(how ='all')
```

```python
all_data = all_data[all_data['Order Date'].str[0:2] !=  'Or']
```

```python
all_data['Quantity Ordered'] =pd.to_numeric(all_data['Quantity Ordered'])

all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])

all_data.head()
```

```python
all_data['Month'] = all_data['Order Date'].str[0:2]
all_data['Month'] = all_data['Month'].astype('int64')
all_data.head()
```

# **3) a. En iyi satış yapılan ay hangisidir? Ne kadar kazanılmıştır?**

```python
all_data['Sales'] =all_data['Quantity Ordered'] * all_data['Price Each']
all_data.head()
```

```python
all_data.groupby('Month').sum()
```

```python
all_data.groupby('Month').sum()['Sales']
```

# **3) b. En çok satın alma yapan şehir hangisidir?**

```python
def get_city(address):
        return address.split(',')[1]

def get_state(address):
        return address.split(',')[2].split(' ')[1]

all_data['City'] = all_data['Purchase Address'].apply(lambda x: get_city(x) + '-' + get_state(x))

all_data.head()
```

```python
all_data.groupby('City').sum()
```

```python
 all_data.groupby('City').sum()['Sales']
```

```python
all_data.head()
```

```python
all_data['Order Date'] = pd.to_datetime(all_data['Order Date']) # sipariş tarihini Sütununu tarih saat nesnesine dönüştürdük

all_data.head()
```

```python
all_data['Hour'] = pd.to_datetime(all_data['Order Date']).dt.hour # saat bilgisini yeni kolonda listeledik

all_data.head()
```

```python
all_data.groupby(['Hour']).count()  # ürünlerin sipariş sayılarını saatlere göre saydı
```
